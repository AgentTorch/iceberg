```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the data
df = pd.read_csv('../data/tech_intensity_simple.csv')

# Create figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Distribution of Hot Tech Ratio', fontsize=16, fontweight='bold')

# 1. Histogram
axes[0, 0].hist(df['hot_tech_ratio'], bins=30, alpha=0.7, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Histogram of Hot Tech Ratio')
axes[0, 0].set_xlabel('Hot Tech Ratio')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].grid(True, alpha=0.3)

# 2. Box plot
axes[0, 1].boxplot(df['hot_tech_ratio'], vert=True)
axes[0, 1].set_title('Box Plot of Hot Tech Ratio')
axes[0, 1].set_ylabel('Hot Tech Ratio')
axes[0, 1].grid(True, alpha=0.3)

# 3. Density plot
axes[1, 0].hist(df['hot_tech_ratio'], bins=30, density=True, alpha=0.7, color='lightcoral', edgecolor='black')
axes[1, 0].set_title('Density Plot of Hot Tech Ratio')
axes[1, 0].set_xlabel('Hot Tech Ratio')
axes[1, 0].set_ylabel('Density')
axes[1, 0].grid(True, alpha=0.3)

# Add a KDE line to the density plot
from scipy.stats import gaussian_kde
kde = gaussian_kde(df['hot_tech_ratio'])
x_range = np.linspace(df['hot_tech_ratio'].min(), df['hot_tech_ratio'].max(), 100)
axes[1, 0].plot(x_range, kde(x_range), 'r-', linewidth=2, label='KDE')
axes[1, 0].legend()

# 4. Cumulative distribution
sorted_data = np.sort(df['hot_tech_ratio'])
y = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
axes[1, 1].plot(sorted_data, y, linewidth=2, color='green')
axes[1, 1].set_title('Cumulative Distribution of Hot Tech Ratio')
axes[1, 1].set_xlabel('Hot Tech Ratio')
axes[1, 1].set_ylabel('Cumulative Probability')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print summary statistics
print("Summary Statistics for Hot Tech Ratio:")
print(f"Mean: {df['hot_tech_ratio'].mean():.4f}")
print(f"Median: {df['hot_tech_ratio'].median():.4f}")
print(f"Standard Deviation: {df['hot_tech_ratio'].std():.4f}")
print(f"Min: {df['hot_tech_ratio'].min():.4f}")
print(f"Max: {df['hot_tech_ratio'].max():.4f}")
print(f"25th Percentile: {df['hot_tech_ratio'].quantile(0.25):.4f}")
print(f"75th Percentile: {df['hot_tech_ratio'].quantile(0.75):.4f}")

# Additional analysis: occupations with highest and lowest hot tech ratios
print("\nTop 10 Occupations with Highest Hot Tech Ratio:")
top_10 = df.nlargest(10, 'hot_tech_ratio')[['O*NET-SOC Code', 'hot_tech_ratio', 'total_technologies', 'hot_technologies']]
print(top_10.to_string(index=False))

print("\nTop 10 Occupations with Lowest Hot Tech Ratio:")
bottom_10 = df.nsmallest(10, 'hot_tech_ratio')[['O*NET-SOC Code', 'hot_tech_ratio', 'total_technologies', 'hot_technologies']]
print(bottom_10.to_string(index=False))
```

```python
# Create figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Distribution of Tech Intensity Score', fontsize=16, fontweight='bold')

# 1. Histogram
axes[0, 0].hist(df['tech_intensity_score'], bins=30, alpha=0.7, color='lightgreen', edgecolor='black')
axes[0, 0].set_title('Histogram of Tech Intensity Score')
axes[0, 0].set_xlabel('Tech Intensity Score')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].grid(True, alpha=0.3)

# 2. Box plot
axes[0, 1].boxplot(df['tech_intensity_score'], vert=True)
axes[0, 1].set_title('Box Plot of Tech Intensity Score')
axes[0, 1].set_ylabel('Tech Intensity Score')
axes[0, 1].grid(True, alpha=0.3)

# 3. Density plot
axes[1, 0].hist(df['tech_intensity_score'], bins=30, density=True, alpha=0.7, color='orange', edgecolor='black')
axes[1, 0].set_title('Density Plot of Tech Intensity Score')
axes[1, 0].set_xlabel('Tech Intensity Score')
axes[1, 0].set_ylabel('Density')
axes[1, 0].grid(True, alpha=0.3)

# Add a KDE line to the density plot
from scipy.stats import gaussian_kde
kde = gaussian_kde(df['tech_intensity_score'])
x_range = np.linspace(df['tech_intensity_score'].min(), df['tech_intensity_score'].max(), 100)
axes[1, 0].plot(x_range, kde(x_range), 'r-', linewidth=2, label='KDE')
axes[1, 0].legend()

# 4. Cumulative distribution
sorted_data = np.sort(df['tech_intensity_score'])
y = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
axes[1, 1].plot(sorted_data, y, linewidth=2, color='purple')
axes[1, 1].set_title('Cumulative Distribution of Tech Intensity Score')
axes[1, 1].set_xlabel('Tech Intensity Score')
axes[1, 1].set_ylabel('Cumulative Probability')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print summary statistics
print("Summary Statistics for Tech Intensity Score:")
print(f"Mean: {df['tech_intensity_score'].mean():.4f}")
print(f"Median: {df['tech_intensity_score'].median():.4f}")
print(f"Standard Deviation: {df['tech_intensity_score'].std():.4f}")
print(f"Min: {df['tech_intensity_score'].min():.4f}")
print(f"Max: {df['tech_intensity_score'].max():.4f}")
print(f"25th Percentile: {df['tech_intensity_score'].quantile(0.25):.4f}")
print(f"75th Percentile: {df['tech_intensity_score'].quantile(0.75):.4f}")

# Additional analysis: occupations with highest and lowest tech intensity scores
print("\nTop 10 Occupations with Highest Tech Intensity Score:")
top_10 = df.nlargest(10, 'tech_intensity_score')[['O*NET-SOC Code', 'tech_intensity_score', 'total_technologies', 'hot_technologies', 'hot_tech_ratio']]
print(top_10.to_string(index=False))

print("\nTop 10 Occupations with Lowest Tech Intensity Score:")
bottom_10 = df.nsmallest(10, 'tech_intensity_score')[['O*NET-SOC Code', 'tech_intensity_score', 'total_technologies', 'hot_technologies', 'hot_tech_ratio']]
print(bottom_10.to_string(index=False))

# Create an additional plot showing the relationship between tech intensity score and other variables
fig2, axes2 = plt.subplots(1, 2, figsize=(15, 6))
fig2.suptitle('Tech Intensity Score Relationships', fontsize=16, fontweight='bold')

# Scatter plot: Tech Intensity Score vs Total Technologies
axes2[0].scatter(df['total_technologies'], df['tech_intensity_score'], alpha=0.6, color='blue')
axes2[0].set_xlabel('Total Technologies')
axes2[0].set_ylabel('Tech Intensity Score')
axes2[0].set_title('Tech Intensity Score vs Total Technologies')
axes2[0].grid(True, alpha=0.3)

# Scatter plot: Tech Intensity Score vs Hot Tech Ratio
axes2[1].scatter(df['hot_tech_ratio'], df['tech_intensity_score'], alpha=0.6, color='red')
axes2[1].set_xlabel('Hot Tech Ratio')
axes2[1].set_ylabel('Tech Intensity Score')
axes2[1].set_title('Tech Intensity Score vs Hot Tech Ratio')
axes2[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Calculate correlation coefficients
print(f"\nCorrelation between Tech Intensity Score and Total Technologies: {df['tech_intensity_score'].corr(df['total_technologies']):.4f}")
print(f"Correlation between Tech Intensity Score and Hot Technologies: {df['tech_intensity_score'].corr(df['hot_technologies']):.4f}")
print(f"Correlation between Tech Intensity Score and Hot Tech Ratio: {df['tech_intensity_score'].corr(df['hot_tech_ratio']):.4f}")
```
